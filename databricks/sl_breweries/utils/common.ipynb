```python
import json
```

```python
def path_exists(path: str) -> bool:
    """
    Check if a path already exists in the storage account.
    
    Args:
        path (str): Path to check.
    Returns:
        bool: True if the path exists, False otherwise.    
    """

    try:
        dbutils.fs.ls(path)
        return True
    except Exception:
        return False
```

```python

def write_bytes_to_path(bytes_content: bytes, dst_path: str) -> None:
    """
    Persist bytes to ADLS path.
    """
    
    tmp_file = f"/tmp/{dst_path.split('/')[-1]}"
    with open(tmp_file, "wb") as f:
        f.write(bytes_content)
    dbutils.fs.cp(f"file:{tmp_file}", dst_path)
```

```python
def write_success_marker(path: str, payload: dict) -> None:
    """
    Create _SUCCESS marker with metadata.
    """
    
    marker_path = f"{path}/_SUCCESS"
    dbutils.fs.put(marker_path, json.dumps(payload), overwrite=True)
```

```python
def read_text(path: str) -> str:
    """
    Read the file's content
    """
    return dbutils.fs.head(path).strip()
```
