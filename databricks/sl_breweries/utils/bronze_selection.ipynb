```python
def is_empty_payload(raw_bytes: bytes) -> bool:
    """
    Stop condition:
    - parse JSON
    - empty list => stop
    """

    try:
        data = json.loads(raw_bytes.decode("utf-8"))
        return isinstance(data, list) and len(data) == 0
    except Exception:
        raise Exception("Could not parse API response as JSON list")
```
