```python
from pyspark.sql import functions as F
```

```python
def build_gold(df_silver):
    """
    Gold table generated by aggregating the silver table by country, state, city, and brewery type.
    Business definition:
      - remove records with encoding issues
      - remove records without an address
    """
    return (
        df_silver
        .filter((F.col("has_proper_encoding") == True) & (F.col("has_address") == True))
        .groupBy("country", "state", "city", "brewery_type")
        .agg(F.countDistinct("id").alias("brewery_count"))
        .withColumn("last_update", F.current_timestamp())
    )
```
